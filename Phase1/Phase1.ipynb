    <h1>
    <b>پیش‌پردازش و آماده‌سازی داده‌ها (۱۵ نمره)</b>
    </h1>
    در این بخش ابتدا داده‌ها را از فایل بخوانید. برای آماده‌سازی متن می‌توانید از کتاب‌خانه‌های آماده استفاده کنید. یکی از کتاب‌خانه‌های معروف برای این کار <a href="https://www.nltk.org/">NLTK</a> است اما در انتخاب روش پیاده‌سازی این بخش مختارید. برای این بخش باید تابع ()prepare_text را تکمیل کنید. این تابع یک متن انگلیسی ورودی گرفته و توکن‌‌های مربوط به آن‌را در قالب یک لیست خروجی می‌دهد. متن ورودی در عمل تایتل یا طرح داستان هر فیلم است. دقت کنید که لیست خروجی شامل تعدادی توکن است که عملیات case folding، stemming و lemmatization روی آن‌ها اجرا شده است. در ضمن علائم نگارشی نباید به عنوان توکن در نظر گرفته شود. در کد زیر یک نمونه ورودی و خروجی نمایش داده شده است. با توجه به نحوه پیاده‌سازی انواع بازگردانی به ریشه قابل قبول است.
</font>
</div>

```python
import pandas as pd
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer

df = pd.read_csv('movies.csv')
data = df.values.tolist()


def prepare_text(raw_text):
    raw_text = raw_text.lower()
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(raw_text)
    lemmatizer = WordNetLemmatizer()
    ps = PorterStemmer()
    for i in range(len(tokens)):
        tokens[i] = ps.stem(tokens[i])
        tokens[i] = lemmatizer.lemmatize(tokens[i])
    return tokens


prepare_text("Edvard was a Runner. He was always running.")
prepare_text("Also, his name is Ali")
prepare_text("wrong,, test.")
```

    این بخش باید توسط خودتان و بدون استفاده از کد آماده پیاده‌سازی شود. ترم‌های موجود در مجموعه دادگان را بر اساس تکرار آن‌ها مرتب کرده و پرتکرارترین آن‌ها را به عنوان stop-words در نظر بگیرید. اینکه چند ترم را به عنوان stop-words در نظر بگیرید به عهده خودتان است.<br>
    با فراخوانی تابع ()get_stop_words لیست stop-words به همراه تعداد تکرار آن‌‌ها خروجی داده می‌شود.<br>

گزارش: برای این بخش پس از پیدا کردن تمام توکن ها و تعداد تکرار آنها سی تا پرتکرار ترین به عنوان استاپ ورد در نظر گرفته شده اند

```python
def get_stop_words():
    dictionary = {}
    title = ""
    plot = ""
    tokens = []
    for i in range(len(data)):
        title = prepare_text(data[i][1])
        plot = prepare_text(data[i][2])
        tokens += title + plot
    for j in range(len(tokens)):
        if dictionary.get(tokens[j]) == None:
            dictionary[tokens[j]] = 0
            continue
        dictionary[tokens[j]] += 1  
    dictionary = dict(sorted(dictionary.items(), key=lambda item: -item[1]))
    items = dictionary.items()
    return list(items)[:30]


stop_words = get_stop_words()
stop_words
```

    <b>ذخیره و فشرده‌سازی نمایه (۲۰ نمره)</b>
    </h1>
    در این بخش باید توانایی ذخیره کردن نمایه و بارگذاری مجدد آن را به سامانه اضافه کنید. ذخیره‌سازی به ۳ روش صورت می‌گیرد. بدون فشرده‌سازی، فشرده‌سازی از روش gamma-code و فشرده‌سازی از روش variable-byte. روش‌های فشرده‌سازی باید توسط خودتان پیاده‌سازی شود.
    برای ذخیره نمایه در فایل نیز از JSON  استفاده کنید.<br>
     بخشی از نمره شما در این قسمت به میزان فشرده‌سازی نمایه اختصاص داده شده است. بنابراین پیاده‌سازی بهینه روش‌های فشرده‌سازی مهم است.

```python
import json
from pathlib import Path

def get_gamma_code(num):
    num += 1
    offset = "{0:b}".format(int(num))
    offset = offset[1:]
    length = ""
    for i in range(len(offset)):
        length += "1"
    length += "0"
    gamma_code = length + offset
    return gamma_code

def get_variable_byte(num):
    binary = "{0:b}".format(int(num))
    if len(binary) < 7:
        while len(binary) < 7:
            binary = "0" + binary
        binary = "1" + binary
        return binary
    if len(binary) == 7:
        return ("1"+ binary)
    remain = binary
    vb = ""
    while len(remain) > 7:
        vb = "0" + remain[len(remain)-7:]
        remain = remain[:len(remain)-7]
    vb = remain + vb
    while len(vb) % 8 != 7:
        vb = "0" + vb
    vb = "1" + vb
    return vb

def get_gamma_string(arr):
    gamma_code = ""
    for num in arr:
        gamma_code += get_gamma_code(num)
    return gamma_code


def get_vb_string(arr):
    vb_code = ""
    for num in arr:
        vb_code += get_variable_byte(num)
    return vb_code

def store_index(path, compression_type):
    if compression_type == "no-compression":
        with open(path, "w") as outfile: 
            json.dump(indexing, outfile)
        print(Path(path).stat().st_size / 1000000, "MB")
        return
    pos_index = {}
    if compression_type == "gamma-code":
        for term in indexing:
            pos_index[term] = [[], []]
            title_pl = indexing[term][0]
            plot_pl = indexing[term][1]
            for i in range(len(title_pl)):
                record = title_pl[i]
                posting_list = record[1]
                gaps = []
                doc_id = record[0] if i < 1 else (title_pl[i][0] - title_pl[i-1][0])
                gaps.append(doc_id)
                gaps.append(posting_list[0])
                if len(posting_list) > 1:
                    for j in range(1,len(posting_list)):
                        gaps.append(posting_list[j] - posting_list[j-1])
                pos_index[term][0].append(hex(int(get_gamma_string(gaps), 2)))
            for i in range(len(plot_pl)):
                record = plot_pl[i]
                posting_list = record[1]
                gaps = []
                doc_id = record[0] if i < 1 else (plot_pl[i][0] - plot_pl[i-1][0])
                gaps.append(doc_id)
                gaps.append(posting_list[0])
                if len(posting_list) > 1:
                    for j in range(1,len(posting_list)):
                        gaps.append(posting_list[j] - posting_list[j-1])
                pos_index[term][1].append(hex(int(get_gamma_string(gaps), 2))) 
    elif compression_type == "variable-byte":
        for term in indexing:
            pos_index[term] = [[], []]
            title_pl = indexing[term][0] 
            plot_pl = indexing[term][1]
            for i in range(len(title_pl)):
                record = title_pl[i]
                posting_list = record[1]
                gaps = []
                doc_id = record[0] if i < 1 else (title_pl[i][0] - title_pl[i-1][0])
                gaps.append(doc_id)
                gaps.append(posting_list[0])
                if len(posting_list) > 1:
                    for j in range(1,len(posting_list)):
                        gaps.append(posting_list[j] - posting_list[j-1])
                pos_index[term][0].append(hex(int(get_vb_string(gaps), 2)))
            for i in range(len(plot_pl)):
                record = plot_pl[i]
                posting_list = record[1]
                gaps = []
                doc_id = record[0] if i < 1 else (plot_pl[i][0] - plot_pl[i-1][0])
                gaps.append(doc_id)
                gaps.append(posting_list[0])
                if len(posting_list) > 1:
                    for j in range(1,len(posting_list)):
                        gaps.append(posting_list[j] - posting_list[j-1])
                pos_index[term][1].append(hex(int(get_vb_string(gaps), 2)))
    with open(path, "w") as outfile: 
            json.dump(pos_index, outfile)
    print(Path(path).stat().st_size / 1000000, "MB")



store_index("gamma.json", "gamma-code")
```

```python
def create_bigram_index():
    bigram_index = {}
    for term in indexing:
        t = "$"
        t += (term + "$")
        for i in range(len(t)-1):
            bigram = t[i:i+2]
            if bigram_index.get(bigram) == None:
                bigram_index[bigram] = [term]
                continue
            bigram_index[bigram].append(term)
    return bigram_index


bigram_index = create_bigram_index()
print(len(bigram_index))
```

```python
import nltk
def get_corrected_text(raw_text):
    if indexing.get(raw_text) != None:
        return raw_text
    query = "$" + raw_text + "$"
    terms = set()
    query_bigrams = set()
    for i in range(len(query)-1):
        bigram = query[i:i+2]
        query_bigrams.add(bigram)
        if bigram_index.get(bigram) == None:
            continue
        terms.update(bigram_index[bigram])
    candidates = []
    for term in terms:
        term_bigrams = set()
        for i in range(len(term)-1):
            term_bigrams.add(term[i:i+2])
        intersect = len(query_bigrams & term_bigrams)
        union = len(query_bigrams) + len(term_bigrams) - intersect
        jaccard = intersect / union
        if jaccard >= 0.2:
            candidates.append(term)
    best_term = ""
    best_distance = 1000
    for term in candidates:
        distance = nltk.edit_distance(raw_text, term, transpositions=False)
        if distance < best_distance:
            best_distance = distance
            best_term = term
    return best_term

print(get_corrected_text("sherlock"))
print(get_corrected_text("preform"))
print(get_corrected_text("herlock"))
```

```python
def precision(actual, predicted):
    intersect = len(list(set(actual) & set(predicted)))
    retrieved = len(predicted)
    return (intersect/retrieved)

precision([1,2], [1, 4, 5])
```

```python
def recall(actual, predicted):
    intersect = len(list(set(actual) & set(predicted)))
    relevent = len(actual)
    return (intersect/relevent)

recall([1,2], [1, 4, 5])
```

```python
def f1_score(actual, predicted):
    p = precision(actual, predicted)
    r = recall(actual, predicted)
    if p+r == 0:
        return 0
    return (2*p*r)/(p+r)


f1_score([1,2], [1, 4, 5])
```

```python
def map_score(actual, predicted):
    temp, cnt = 0, 0
    for i in range(len(predicted)):
        if predicted[i] in actual:
            temp += precision(actual, predicted[:i+1])
            cnt += 1
    if cnt == 0:
        return 0
    return temp / cnt

map_score([1,2], [1, 4, 5])
```

```python
import math
def ndcg_score(actual, predicted):
    ideal = 1
    for i in range(1,len(actual)):
        ideal += (1/math.log(i+1, 2))
    dcg = 1 if predicted[0] in actual else 0
    for i in range(1, len(predicted)):
        score = 1 if predicted[i] in actual else 0
        dcg += (score/math.log(i+1, 2))
    return dcg/ideal

ndcg_score([1,2], [1, 4, 5])
```
